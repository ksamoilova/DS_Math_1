```python
df_boston = pd.read_csv(boston['filename'], skiprows=1)
df_boston.head()
```

Нарисовать два дома в 3-мерном пространстве (атрибуты выбрать самостоятельно)

```python
fig = plt.figure(figsize=(10, 10))
ax = fig.gca(projection='3d')

x = np.array(df_boston['RM'])
y = np.array(df_boston['AGE'])
z = np.array(df_boston['MEDV'])

ax.plot([0,x[0]], [0,y[0]], [0,z[0]])
ax.plot([0,x[1]], [0,y[1]], [0,z[1]])


ax.scatter(x[0], y[0], z[0], c='b', marker='o')
ax.scatter(x[1], y[1], z[1], c='r', marker='o')

ax.set_xlabel('RM')
ax.set_ylabel('AGE')
ax.set_zlabel('MEDV')

plt.show()
```

Найти косинус угла между векторами x1 и x2, где x1 и x2 - это два дома из 2ух атрибутов, полученных на этапе 2.

Найти расстояние между точками x1 и x2.

```python
print (x1)
print (x2)
```

```python
a = math.sqrt(x1[0]**2 + x1[1]**2)
b = math.sqrt(x2[0]**2 + x2[1]**2)
cos = np.dot (x1, x2) /(a*b)
print ('Косинус угла между векторами х1 и х2 ', cos)
```

```python
dist = math.sqrt ((x2[0] - x1[0])**2 + (x2[1] - x1[1])**2)
print ('Расстояние между точками х1 и х2 ', dist)
```
